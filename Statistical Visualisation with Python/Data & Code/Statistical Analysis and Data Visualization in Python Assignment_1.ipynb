# Data

### About the Data

The synthetic dataset generated consists of 1000 samples and includes the following variables: Age, Height, Weight, Gender, and Income. This dataset was created using specified statistical distributions to simulate a diverse population and to facilitate various analyses. Below is a detailed description of each variable and the methodology used for generating the data:

1. **Age**:
   - **Description**: Represents the age of individuals in years.
   - **Generation Method**: Age is generated using a normal distribution with a mean of 35 years and a standard deviation of 10 years. The values are then converted to integers.
   - **Range**: 2 to 73 years.
   - **Distribution**: The distribution is roughly symmetric around the mean, indicating a typical spread of ages with a few younger and older individuals.

2. **Height**:
   - **Description**: Represents the height of individuals in centimeters.
   - **Generation Method**: Height is generated using a normal distribution with a mean of 170 cm and a standard deviation of 15 cm. The values are rounded to two decimal places.
   - **Range**: 125.89 cm to 217.90 cm.
   - **Distribution**: Heights follow a normal distribution centered around the mean, with most values falling within a typical human height range and a few outliers at the extremes.

3. **Weight**:
   - **Description**: Represents the weight of individuals in kilograms.
   - **Generation Method**: Weight is generated using a normal distribution with a mean of 70 kg and a standard deviation of 10 kg. The values are rounded to two decimal places.
   - **Range**: 39.80 kg to 109.26 kg.
   - **Distribution**: Weights are normally distributed around the mean, with a moderate spread and a few outliers indicating very light or very heavy individuals.

4. **Gender**:
   - **Description**: Represents the gender of individuals, categorized as 'Male' or 'Female'.
   - **Generation Method**: Gender is randomly assigned with a 50% probability for each category.
   - **Distribution**: The dataset has an approximately equal number of males and females.

5. **Income**:
   - **Description**: Represents the annual income of individuals in dollars.
   - **Generation Method**: Income is generated using a normal distribution with a mean of 50,000 and a standard deviation of 15,000. The values are rounded to two decimal places.
   - **Range**: 4,885.51 to 98,646.39.
   - **Distribution**: Incomes exhibit a wider spread compared to other variables, indicating a diverse set of economic backgrounds among the individuals. There are several outliers, reflecting significant variability in income levels.

```python
import numpy as np
import pandas as pd

# Setting the seed for reproducibility
np.random.seed(42)

# Number of samples
num_samples = 1000

# Generating data
age = np.random.normal(35, 10, num_samples).astype(int)
height = np.random.normal(170, 15, num_samples).round(2)
weight = np.random.normal(70, 10, num_samples).round(2)
gender = np.random.choice(['Male', 'Female'], num_samples)
income = np.random.normal(50000, 15000, num_samples).round(2)

# Creating DataFrame
data = pd.DataFrame({
    'Age': age,
    'Height in cm': height,
    'Weight in Kg': weight,
    'Gender': gender,
    'Income': income
})

# Displaying the dataset
display(data.head())
display(data.tail())
```

# Descriptive Statistics

```python
# Task 2: Calculating basic descriptive statistics
descriptive_stats = data[['Age', 'Height in cm', 'Weight in Kg', 'Income']].describe().T
descriptive_stats['variance'] = data[['Age', 'Height in cm', 'Weight in Kg', 'Income']].var()

# Calculating median for Age, Height, Weight, and Income
medians = data[['Age', 'Height in cm', 'Weight in Kg', 'Income']].median()

# Adding median to the descriptive statistics
descriptive_stats['median'] = medians

# Calculating the mode for Gender
gender_mode = data['Gender'].mode()[0]

# Displaying the results
print("Descriptive Statistics:")
display(descriptive_stats)
print("\nMode for Gender:", gender_mode)
```

# Inferential Statistics (T test)

```python
from scipy import stats

# Splitting the income data based on gender
male_income = data[data['Gender'] == 'Male']['Income']
female_income = data[data['Gender'] == 'Female']['Income']

# Performing a t-test to see if there is a significant difference in Income between Male and Female
t_stat, p_value = stats.ttest_ind(male_income, female_income)

print(f"t-statistic: {t_stat}, p-value: {p_value}")
```

### Boxplot Components

1. **Median (Q2)**: The line inside the box represents the median (50th percentile) of the data.
2. **Interquartile Range (IQR)**: The box represents the IQR, which contains the middle 50% of the data. It spans from the first quartile (Q1, 25th percentile) to the third quartile (Q3, 75th percentile).
3. **Whiskers**: The lines extending from the box (whiskers) represent the range within 1.5 times the IQR from Q1 and Q3. Points outside this range are considered potential outliers.
4. **Outliers**: Individual points beyond the whiskers are plotted as dots and are considered potential outliers.

## 6. Explain the results of the t-test. What does the p-value indicate about the difference in Income between Male and Female?

The t-test was performed to determine if there is a statistically significant difference in income between males and females. Here are the key results from the t-test and their interpretation:

### T-test Results
- **t-statistic**: -1.967
- **p-value**: 0.049

### Interpretation of the T-test Results

#### 1. **t-statistic**:
- The t-statistic value of -1.967 indicates the difference in means between the two groups (males and females) expressed in terms of the standard error. 
- The negative sign indicates that the mean income for males is slightly less than that for females in this dataset.

#### 2. **p-value**:
- The p-value of 0.049 is a measure of the probability that the observed difference in means (or a more extreme difference) would occur if there were no actual difference in the population (null hypothesis is true).
- A commonly used threshold for statistical significance is 0.05. If the p-value is less than or equal to this threshold, we reject the null hypothesis.

### Conclusion Based on the p-value:
- **p-value = 0.049**: Since the p-value is slightly less than the significance level of 0.05, we reject the null hypothesis.
- This indicates that there is sufficient evidence to suggest a statistically significant difference in income between males and females at the 5% significance level.

### Practical Implications:
- The result implies that, based on our synthetic dataset, gender has a statistically significant effect on income.
- However, it is important to note that "statistically significant" does not necessarily mean "practically significant." The actual difference in mean incomes might be small, and other factors could also influence income.
- Further analysis would be required to determine the magnitude of this difference and to understand the underlying causes. Additionally, it's important to ensure that the dataset is representative and that other potential confounding variables are considered.

### Summary:
- The t-test result indicates a statistically significant difference in income between males and females, with the p-value of 0.049 just below the commonly used threshold of 0.05.
- While the statistical significance suggests a real difference in income based on gender, the practical significance and underlying factors would require further investigation.
